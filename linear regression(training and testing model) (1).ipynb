```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #scikit learn
from sklearn.model_selection import train_test_split
```

```python
studytime=np.array([20,50,32,65,23,43,10,5,22,35,29,5,56]).reshape(-1,1)
scores=np.array([56,83,47,93,47,82,45,78,55,67,57,4,17]).reshape(-1,1)
studytime
```

In this model, i've split the given data and i have taken 80% of data for training and 20% data for testing which is the ratio commonly used to know how accurately the model can predict values. If you have enough data you can follow this ration or else choose the ratio according to your data. It splits the whole data & trains the model with the given ratio of train data(eg. 80%) and test the data with the test ratio specified(20%) which gives the accurate value.

```python
studytime_train,studytime_test,scores_train,scores_test=train_test_split(studytime,scores,test_size=0.2)
```

```python
model=LinearRegression()
model.fit(studytime_train,scores_train)
model.score(studytime_test,scores_test)
```

```python
plt.scatter(studytime_train,scores_train)
plt.plot(np.linspace(0,70,100).reshape(-1,1),model.predict(np.linspace(0,70,100).reshape(-1,1)),'r')
```
